문제 1) 무한 제너레이터에서 조건 만족하는 첫 N개 추출
1부터 시작하는 자연수 중 3의 배수만 필터링한 뒤, 처음 10개를 출력하세요.  
[출력]  
3, 6, 9, ..., 30

````python
import itertools

gen = (x for x in itertools.count(1) if x % 3 == 0)
print(list(itertools.islice(gen, 10)))
````

문제 2) 한 줄에서 단어 빈도수 계산

```
# 예: line = "data science is data"
```

문제 5) sorted의 key 인자 활용
다음 리스트에서 길이순 정렬 후, 알파벳 역순으로 2차 정렬하세요.  
["banana", "apple", "kiwi", "avocado", "fig"]  

[출력]  
['fig', 'kiwi', 'apple', 'banana', 'avocado']

````python
words = ["banana", "apple", "kiwi", "avocado", "fig"]
print(sorted(words, key=lambda x: (len(x), -ord(x[0]))))
````

문제 6) groupby로 구간 그룹핑
[75, 82, 90, 66, 72, 88, 55] 점수를 10점 단위로 그룹핑해서 출력하세요.    
```
{
  '50~59': [55],
  '60~69': [66],
  ...
}

```

````python
from collections import defaultdict

scores = [75, 82, 90, 66, 72, 88, 55]
grouped = defaultdict(list)
for s in scores:
    k = f'{s//10*10}~{s//10*10+9}'
    grouped[k].append(s)
print(dict(grouped))


````

````python
from collections import defaultdict

data = [('a', 3), ('b', 1), ('a', 2)]
d = defaultdict(int)
for k, v in data:
    d[k] += v
print(dict(d))
````

문제 15) 리스트를 n개씩 슬라이싱
[1,2,3,4,5,6,7] 리스트를 크기 3으로 나눈 하위 리스트를 만드세요.  

[출력]  
[[1,2,3], [4,5,6], [7]]  

````python
def chunked(seq, n):
    return [seq[i:i+n] for i in range(0, len(seq), n)]

print(chunked([1,2,3,4,5,6,7], 3))
````

문제19) 조건에 맞는 누적합  

리스트 [1, 2, 3, 4, 5]의 누적합 중 10 이하까지만 출력하세요.  
[출력]   
 1, 3, 6, 10

````python
from itertools import accumulate, takewhile

nums = [1, 2, 3, 4, 5]
for x in takewhile(lambda x: x <= 10, accumulate(nums)):
    print(x)

````

문제21) 최단 거리 순회 (브루트포스 + 순열)  
도시 A, B, C, D 간 거리 행렬이 주어질 때, 모든 순열을 돌며 최단 거리 경로를 구하세요.
(여행자 문제의 단순 브루트포스 버전)

````python
from itertools import permutations

cities = ['A', 'B', 'C', 'D']
distance = {
    ('A', 'B'): 10, ('A', 'C'): 15, ('A', 'D'): 20,
    ('B', 'C'): 35, ('B', 'D'): 25,
    ('C', 'D'): 30
}
# 대칭이므로 역방향도 포함
for (a, b), v in list(distance.items()):
    distance[(b, a)] = v

min_path = None
min_dist = float('inf')

for path in permutations(cities):
    dist = 0
    for i in range(len(path) - 1):
        dist += distance[(path[i], path[i + 1])]
    if dist < min_dist:
        min_dist = dist
        min_path = path

print("최단 경로:", min_path)
print("총 거리:", min_dist)

````

문제22) 효율적인 필터링 (filterfalse, dropwhile)  
리스트에서 짝수만 제거하거나 특정 조건 이전까지 생략하려면?

````python
from itertools import filterfalse, dropwhile

nums = [1, 2, 3, 4, 5, 6]

# 짝수 제거 (filterfalse: 조건이 False인 값만 통과)
filtered = list(filterfalse(lambda x: x % 2 == 0, nums))
print("짝수 제거:", filtered)

# 3보다 작을 때까지 생략 (dropwhile)
dropped = list(dropwhile(lambda x: x < 3, nums))
print("3보다 작은 값 생략:", dropped)
````

문제 23) zip과 map을 이용한 다차원 행렬 전치
2차원 리스트를 전치(transpose)하세요.  
입력: [[1,2,3],[4,5,6]]  
출력: [[1,4],[2,5],[3,6]]  

````python
matrix = [[1,2,3],[4,5,6]]
print(list(map(list, zip(*matrix))))
````
