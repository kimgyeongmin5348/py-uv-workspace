[문제2]   
문자열을 입력받아 각 문자들로 구성된 리스트로 반환하는 make_list 함수를 정의하세요.  
make_list("abcd")  

[출력결과]

['a', 'b', 'c', 'd']


```python
answer=[]
def make_list(str):
    for i in str:
        answer.append(i)

make_list('abcd')
print(answer)
```

[문제3]   
숫자로 구성된 하나의 리스트를 입력받아, 짝수들을 추출하여 리스트로 반환하는 pickup_even 함수를 구현하세요.   

pickup_even([3, 4, 5, 6, 7, 8])

[출력결과]   
[4, 6, 8]

```python
def pickup_even(list):
    answer =[]
    for i in list:
        if i%2==0:
            answer.append(i)
    print(answer)

pickup_even([3,4,5,6,7,8])
```

[문제4]  
콤마가 포함된 문자열 숫자를 입력받아 정수로 변환하는 convert_int 함수를 정의하세요.

convert_int("1,234,567")

[출력결과]   
1234567

```python
def convert_int(str):
    for i in str:
       if i.isdigit():
            print(i,end="")

convert_int("1,234,567")
```

[문제4]  
세 개의 숫자를 입력받아 가장 큰수를 출력하는 print_max 함수를 정의하라. 단 if 문을 사용해서 수를 비교하세요.   
max(10, 2, 5)

[출력결과]  
10

```python
def max(a,b,c):
    if a > b and a > c:
        print(a)
    elif b>a and b>c:
        print(b)
    else:
        print(c)

max(10,2,5)
```

[문제5]    
2 와 5의 배수가 아닌 갯수와 합계를 구하는 함수를 구현하세요.    

resultr(1, 3, 5, 8, 9, 11, 15, 19, 18, 20, 30, 33, 31)  

[출력결과]    
2 와 5의 배수가 아닌 갯수 : 7    
2 와 5의 배수가 아닌 합계 : 107   

==============================================  
result(5,2,7,9,10,8,6)     

[출력결과]    
2 와 5의 배수가 아닌 갯수 :  2
2 와 5의 배수가 아닌 합계 :  16



```python
def resultr(*arg):
    c = 0
    d = 0
    for i in arg:
        if i%2!=0 and i%5!=0:
            c += 1
            d += i

    print(f"2와 5의 배수가 아닌 갯수 : {c}")
    print(f"2와 5의 배수가 아닌 합계 : {d}")       


resultr(1, 3, 5, 8, 9, 11, 15, 19, 18, 20, 30, 33, 31)     
resultr(5,2,7,9,10,8,6)    
```

```python
# 문제7)
def introduce(name, /, age, *, country):
    return f"{name}, {age} years old from {country}"
```

```python
# 문제8)
def f(x, *args, y, **kwargs):
    return x, args, y, kwargs
```

```python
# 문제9)
def display(*args, **kwargs):
    print("ARGS:", args)
    print("KWARGS:", kwargs)
```

```python
# 문제10)
def config(*, debug=False, verbose=False):
    return f"Debug={debug}, Verbose={verbose}"
```

```python
# 문제11)
def info(name, /, *, age, country):
    return f"{name}, {age}, {country}"
```

```python
# 문제12)
def combine(a, /, b, *, c):
    return a + b + c
```
