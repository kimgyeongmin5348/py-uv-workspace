## 함수(function)
 - 함수: 특정 목적을 가진 코드의 집합이며 독립적으로 호출될 수 있는 기능이다.
 - 함수 구조
  ```
    def 함수이름(parameter_1, parameter_2) :
        수행할 문장
        return 결과값
  ```
 - 함수 호출
  ```
     함수이름(argument_1, argument_2 )
  ```
   parameter : 인자, argument : 인수
