```
  연산자 (operator) : 어떠한 기능을 수행하는 기호(+, -, *, /, %, >, <, >=, <=, ==, != ...)
  피연산자(operand) : 연산자의 작업 대상 ( 변수, 상수, 리터널, 수식...)
 
  연산자의 종류
 1. 산술연산자 : +, -, *, /,//, %
 2. 비교연산자 : >, <, >=, <=, ==, !=
 3. 논리연산자 : and, or, not
 4. 대입연산자 : =, +=, -=, *=, /=,...
 5. 식별연산자 : is, is not
```





````python
print(10 / 4) # 실수 나눗셈을 수행
print(10 // 4) # 정수 나눗셈을 수행


print(2 ** 4) # 2의 4제곱


# %는 정수나눗셈의 나머지결과를 얻는 연산자입니다.
print(4 % 3)
print(27 % 5)


print("-" * 40)


n1 = int(input("첫번째 정수: "))
n2 = int(input("두번째 정수: "))


print(n1, "과", n2, "의 나눗셈의 몫은", n1//n2,
      "이고, 나머지는", n1%n2, "입니다.")
````

# 식별 연산자 (Identity Operators)
- 식별 연산자는 두 객체가 같은 객체인지를 확인한다.
- is : 두 객체가 같은 객체인지 확인
- is not 두 객체가 다른 객체인지 확인
- 결과 : True, False


````python
a = [1, 2, 3]
b = a
c = [1, 2, 3]
print(a is b)  # True (a와 b는 같은 객체)
print(a is c)  # False (a와 c는 같은 값을 가지지만 다른 객체)
````

#   멤버 연산자 (Membership Operators)
- 멤버 연산자는 시퀀스 내에 값이 존재하는지 확인한다.
- in : 특정값이 멤버의 값인지 확인
- in not : 특정값이 멤버의 값인지 아닌지 확인
- 결과 : True, False


````python
a = [1, 2, 3]
print(2 in a)  # True
print(4 not in a)  # True
````
