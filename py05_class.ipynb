````python
class Person:
    def __init__(self, name, age, job):
      self.name = name  # self.멤버변수 = 매개변수
      self.age = age
      self.job = job


    def eat(self):
        print('eat')


    def walk(self):
        print('walk')

````

````python
hong = Person('홍길동',30,'학생')
hong.eat()
hong.walk()
print(hong.name, hong.age, hong.job,'\n')


lee = Person('이영희', 25, '개발자')
lee.eat( )
lee.walk()
print(lee.name, lee.age, lee.job)
````

````python
class FourCal:


  def __init__(self, a, b):
    print('init')
    self.a = a
    self.b = b


  def add(self):
    # 클래스의 멤버 변수를 메소드를 호출해서 사용할 수 있다.
    return self.a + self.b


  def sub(self):
    return self.a - self.b


  def mul(self):
    return self.a * self.b
  def div(self):
    return self.a // self.b

````

````python
fc = FourCal(4,2)   # 생성자는 객체의 초기값을 할당한다.
````

````python


print(fc.add())
print(fc.sub())
print(fc.mul())
print(fc.div())
````

````python
# @을 붙이면 데코레이트(decorator) 이다.
@x
def s(a):
    return a
````

````python
# s() 함수에 @x 데코레이터가 선언되어 있지 않을때 호출하는 방법
print(x(s)(5))

# s() 함수에 @x 데코레이터가 선언되어 있을때 호출하는 방법
print(s(5))
````

### classmethod
- 첫 번째 인자로 클래스(cls)를 받는다.
- 클래스를 이용해 무언가를 해야할때 사용한다.
- 클래스 변수에 접근하거나 다른 클래스 메소드를 호출할 수 있다.
- 일반적으로 팩토리 메서드 등에 자주 사용된다.


````python
class B:
  x=3          # 클래스 변수
  @classmethod
  def aa(cls):  # classmethod는 관례상 cls를 사용한다.
    print('python')


  def bb(self):
    print('instance')

````

````python
B.x # 클래스.클래스 변수
````

````python
B.aa()  # 클래스.클래스메소드
````

````python
ab = B()
B.bb(ab)
````

````python
class B:
    x = 1

    @classmethod
    def aa(cls):
        cls.x = x
    
````

````python
b=B()
````

````python
b.x # b인스턴스에 x가 없으므로 B의 클래스의 x를 가져온다.
````

````python
b.x = 5 # b인스턴스에 x변수를 생성한다.
````
