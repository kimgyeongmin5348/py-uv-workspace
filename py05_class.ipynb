````python
class FourCal:


  def __init__(self, a, b):
    print('init')
    self.a = a
    self.b = b


  def add(self):
    # 클래스의 멤버 변수를 메소드를 호출해서 사용할 수 있다.
    return self.a + self.b


  def sub(self):
    return self.a - self.b


  def mul(self):
    return self.a * self.b
  def div(self):
    return self.a // self.b

````

### staticmethod
- 첫 번째 인자로 self, cls를 받지 않는다.
- 클래스나 인스턴스와 관계없는 일반적인 유틸리티 함수를 정의할때 사용한다.
- 독립적인 기능이지만 클래스 안에 넣는 것이 구조적으로 더 나을때 사용한다.


| 항목         | classmethod             | staticmethod |
| ---------- | ----------------------- | ------------ |
| 첫 번째 인자    | `cls` (클래스)             | 없음           |
| 클래스 변수 접근  | 가능                      | 불가능          |
| 인스턴스 변수 접근 | 불가능                     | 불가능          |
| 주 용도       | 클래스 상태를 사용하는 메서드        | 유틸리티 함수 정의   |
| 호출 방식      | 클래스.메서드() 또는 인스턴스.메서드() | 동일           |


````python
class B:
  x = 1


  def xx(self, x): # 인스턴스 메소드
    self.x = x


  @classmethod
  def yy(cls, x):  # 클래스 메소드
    cls.x = x


  @staticmethod
  def zz(a, b):  # 스테이틱 메소드
    print(a+b)


  @staticmethod  # 스테이틱 메소드
  def kk():
    print(type(x))
````

````python
B.yy(10)
````

````python
B.x
````

````python
B.zz(3,4)
````

````python
B.kk()
````

````python
class Book:
    def __init__(self,title=None, author=None, page=None):  # 밑에 __ 를 사용하는 이유는 private 처리를 하기 위해서.
        self.__title = title
        self.__author = author
        self.__page = page

    def getTitle(self):
        return self.__title
    
    def setTitle(self, title):
        self.__title = title
    
    def getAuthor(self):
        return self.__author
    
    def settAuthor(self,author):
        self.__author = author

    def getPage(self):
        return self.__page
    
    def setPage(self, page):
        self.__page = page

    def toString(self):
        return f"{self.__title} 저자는 {self.__author}이고 {self.__page}페이지로 구성됨"
    
    
````

클래스 메서드로 인스턴스 개수 세기

````python
class Dog:
    count = 0
    def __init__(self, name):
        self.name = name
        Dog.count = Dog.count + 1

    @classmethod
    def getCount(cls):
        return cls.count
    
    @staticmethod
    def is_valid_name(name):
        # name 의 클래스가 문자열이고, 문자열 길이는 0보다 크면 된다.
        return isinstance(name,str) and len(name) > 0
    
````

````python
dog1 = Dog('Max')
# print(Dog.getCount())
print(dog1.getCount())

dog2 = Dog('coco')
# print(Dog.getCount())
print(dog2.getCount())
````

````python
print(Dog.is_valid_name('초코'))
print(Dog.is_valid_name(''))
print(Dog.is_valid_name(123))
````

 - python의 모든 객체는 object를 상속받는다.
 - python3부터는 (object)는 생략 가능하다. 즉 지금까지 class를 정의할때 object를 내부적으로 상속하고 있었다.

````python
# 부모클래스
class Parent:
  x = 'parent x'  #클래스 변수
  y = 'parent y'  #클래스 변수


  @classmethod
  def output(cls):
    return cls.x


  def test(self):
     return 10
  

#자식클래스
class Child(Parent):
  x = 'child x' # 클래스 변수


  def display(self):
    print('display')
    print(Parent.output())
    print(Child.output())
    print(Parent.test(self))
    print(Child.test(self))

````

````python
print(Parent.x)
print(Parent.y,'\n')
print(Child.x)
print(Child.y)
````

````python
c = Child()
c.display()
````

## overriding(오버라이딩)
부모가 가진 기능을 변경하는 것을 overriding이라고 한다.  
double underbar 같 분은 것들은 overriding해서 객체의 기능을 바꿀 수 없다.


````python
# 부모 클래스 정의
class Animal:
    def __init__(self, name):
        self.name = name


    def make_sound(self):
        # 자식 클래스들이 반드시 재정의(오버라이딩)하도록 유도하는 부모의 기본 메서드
        return "알 수 없는 동물의 소리입니다."


# 자식 클래스 1 (Dog)
class Dog(Animal):
    def make_sound(self):
        # 부모의 make_sound 메서드를 자식에 맞게 완전히 새롭게 덮어씀 (오버라이딩)
        return "멍멍!"


# 자식 클래스 2 (Cat)
class Cat(Animal):
    def make_sound(self):
        # 부모의 make_sound 메서드를 자식에 맞게 완전히 새롭게 덮어씀 (오버라이딩)
        return "야옹~"
````

````python
aniaml = Animal('코코')
print(aniaml.make_sound())

dog = Dog('강아지')
print(dog.make_sound())

cat = Cat('고양이')
print(cat.make_sound())
````

````python
from abc import ABC, abstractmethod


class Shape(ABC):


    def __init__(self, width, height):
        self.width = width
        self.height = height


    @abstractmethod
    def area(self):
      pass    




class Rect(Shape):
   
    def __init__(self, width, height ):
        super().__init__(width, height)


    def area(self):
      return self.width * self.height  


   
class Triangle(Shape):
   
    def __init__(self, width, height ):
        super().__init__(width, height)


    def area(self):
      return self.width * self.height * 0.5
````

````python
# s = Shape(3,2) <- 실행 불가능
````

````python
r= Rect(10,5)
print(f"사각형의 넓이 : {r.area()}")

t = Triangle(30,10)
print(f"삼각형의 넓이 : {t.area()}")
````
